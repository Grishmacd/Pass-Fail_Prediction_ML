```python

from google.colab import files
files.upload()

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

df = pd.read_csv("file_pass_fail.csv")

df['message'] = df['subject1'].astype(str)
df['label'] = df['Pass'].map({'yes':1, 'no':0})


X_train, X_test, y_train, y_test = train_test_split(
    df['message'], df['label'], test_size=0.2, random_state=42
)

vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

y_pred = model.predict(X_test_tfidf)
print("Accuracy:", accuracy_score(y_test, y_pred))

def predict_message(mark):
    return "yes" if model.predict(vectorizer.transform([str(mark)]))[0] == 1 else "no"

print(predict_message(40))
print(predict_message(15))
```
